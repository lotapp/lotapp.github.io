### 1.3.通过arange和reshape创建数组

`range()`返回的是`range`对象，而 **`np.arange()`返回的是`numpy.ndarray() `**

两者都可以进行迭代，而`np.nrange`还是一个序列，可被当做向量使用

`range()`不支持步长为小数, **`np.arange()`支持步长为小数**

如果这些基础类型还不是很清楚，可以看我之前写的文章：<a href="https://www.cnblogs.com/dunitian/p/4822808.html#ai" target="_blank">https://www.cnblogs.com/dotnetcrazy/p/9155310.html</a>

---

**多维数组创建**：（reshape的 **行x列 != 元素个数** 就会报错）

创建3行5列的数组：`np.arange(15).reshape((3,5))`

告诉它多少列，自动求行：`np.arange(20).reshape((-1,5))`

告诉它多少行，自动求列：`np.arange(20).reshape((4,-1))`

```python
help(np.arange)
```

```python
list(range(5))
```

```python
np.arange(5)
```

```python
type(range(1,10))
```

```python
# 用法和range差不多
array1 = np.arange(1,10) # [1,10)

print(array1)

type(array1)
```

```python
# 可以直接迭代
for i in array1:
    print(i,end=' ')
```

```python
# Python里面的range不支持步长为小数
range(1,10,0.5)
```

```python
# 支持步长为小数
np.arange(1,10,0.5) # step为0.5
```

```python
################### 开始变化 ########################
```

```python
help(array1.reshape)
```

```python
# 创建2维数组，参数传元组比较合适（官方案例也是元组）
np.arange(1,10,0.5).reshape((2,9))
```

```python
# 3行5列
np.arange(15).reshape((3,5))
```

```python
# 告诉它多少列，自动求行
np.arange(20).reshape((-1,5))
```

```python
# 告诉它多少行，自动求列
np.arange(20).reshape((4,-1))
```

```python
# reshape的 **行x列 != 元素个数** 就会报错
np.arange(20).reshape(3,7)
```

### 2.2.多维数组的切片用法

**`array_name[行切片,列切片]`** eg：`test_array[1:3,1:3]`==> 取索引为1,2的行；取索引为1,2的列

基础用法之前讲过，可以回顾一下：<a href="https://www.cnblogs.com/dotnetcrazy/p/9155310.html#1.7.列表切片" target="_blank">https://www.cnblogs.com/dotnetcrazy/p/9155310.html#1.7.列表切片</a>

```python
# 先用熟悉的一维数组
array1 = np.arange(100)

print(array1)
```

```python
# 全部取出来
array1[::]
```

```python
# 从索引为1～最后
array1[1::]
```

```python
################### 切片开始 ########################
```

```python
# 从第二个元素取到最后,step为10
array1[1::10]
```

```python
# [1,5) 从下标1取到下标4,下标为5的取不到
array1[1:5]
```

```python
# 取下标为1～倒数第二个元素（end_index取不到）
array1[1:-1]
```

```python
# 取前10个
array1[:10]
```

```python
# 取最后10个，可以理解为从索引-5的元素到最后
array1[-10:]
```

```python
# 前11~20（eg：第二页）
array1[10:20]
```

```python
# 取80~90（eg:倒数第二页）
array1[-20:-10]
```

```python
# 前20个数，每两个取一个（eg:隔行换样式）
array1[:20:2]
```

```python
# 所有数每10个取一个（eg:test的时候十里挑一）
array1[::10]
```

```python
################### 二维数组 ########################
```

```python
# 生成一个多维数组
test_array = np.arange(15).reshape(3,5)

print(test_array)
```

```python
# [行切片，列切片]
# 行：取第2行到最后，列：取第3列到最后
test_array[1:,2:]
```

```python
# 所有行，最后一列
test_array[:,-1:]
```

```python
# 所有行，索引[0,2)列
test_array[:,:2]
```

```python
# 索引1的行，索引1的列
test_array[1:2,1:2]
```

```python
# 上面的等同于test_array[1][1]
print(test_array[1,1])
```

```python
# 取索引为1,2的行；取索引为1,2的列
test_array[1:3,1:3]
```

### 2.5.eye创建单位矩阵

先普及一个数学基础：**任何`矩阵` x `单位矩阵` 都等于其`本身`**

单位矩阵是个方阵（行列相等），从左上角到右下角的对角线（称为主对角线）上的元素均为1。其他全都为0，eg:

\begin{bmatrix}
1&0&0 \\
0&1&0 \\
0&0&1
\end{bmatrix}

用 **`np.eye()`** 来定义(eye：眼睛)

扩展：`np.eye(rows,columns=rows)`

```python
help(np.eye)
```

```python
# 定义一个2行的单位矩阵（列默认和行一致）
np.eye(2)
```

```python
array1 = np.eye(3,dtype=int)

print(array1)
type(array1)
```

```python
# 定义一个5行5列的单位矩阵
np.eye(5)
```

```python
################### 扩展部分 ########################
```

```python
# 可以创建np.eye(rows,columns=rows)
# 但是这种就不是单位矩阵了（单位矩阵是个方阵）
np.eye(2,3)
```
